### 2. Estado del Arte.
La visión artificial puede ser entendida como un análisis de un mundo tridimensional, posiblemente dinámico, en donde mediante la adquisición de la luz que emite éste mundo, se pueden crear imágenes bidimensionales que permitan la identificación y clasificación de tales propiedades. De esta manera, para las aplicaciones en visión artificial es crítico el uso de campos científicos y tecnológicos como la óptica, la electrónica, la geometría y algorítmica, los sistemas de computación, entre otros. La visión artificial tiene un potencial de aplicación muy grande debido a la cantidad de datos que pueden ser obtenidos, procesados y analizados con sus métodos, permitiendo así crear modelos de visión artificial para muchos escenarios; por lo tanto es importante entender la estructura básica de un proceso de visión artificial. Un sistema de visión artificial está compuesto de manera general por:

* Adquisición: Etapa en que se adquieren los datos del (los) objeto(s) a estudiar, mediante la captura de imágenes de éste(os).
* Pre-procesamiento: Etapa en la que se realiza el tratamiento (en caso de ser necesario) sobre las imágenes obtenidas, con el fin de optimizar la identificación de características de interés, relevantes para el objetivo del proyecto específico. Algunas técnicas usadas en esta etapa son transformaciones de color, posición, iluminación, nivel de detalle, entre otros. 
* Segmentación: Etapa en la que se busca realizar una diferenciación de los elementos de interés en la imagen de aquellos que no lo son, de esta manera los métodos aplicados en esta etapa son función de la importancia de los colores, las formas, la iluminación y otras características presentes en las imágenes de estudio. 
* Extracción de características: una vez diferenciados los elementos de interés en las etapas anteriores, se procede a realizar un aislamiento de éstos de manera completa con el fin concentrar el análisis en ellos exclusivamente. 
* Reconocimiento de patrones: Etapa en la que se realiza una comparación de las características que presentan los elementos de interés presentes en todas las imágenes que están siendo estudiadas, buscando encontrar un patrón de comportamiento similar que permita realizar deducciones, que normalmente son el objetivo principal de todo el proyecto de visión artificial que se haya estructurado para un caso específico. 

En la actualidad, aunque tal vez no sean formalmente reconocidos como procesos en visión artificial, todas estas etapas suelen ser usadas con mucha frecuencia en todo tipo de estudios, como el que nos ocupa en este proyecto en específico: las aplicaciones médicas. Los estudios por biometría, como se mencionó previamente, se ocupan de la identificación de características presentes en componentes específicos del cuerpo humano. “En el caso de el reconocimiento de iris, son las estructuras distintivas del iris humanos las que son usadas para hacerlo.” (Nanavati et al. 2002 citado por Aslam et. al, 2009 https://www.ncbi.nlm.nih.gov/pmc/articles/PMC2659699/) Lo que hace que el reconocimiento de Iris, hoy por hoy se haya convertido en uno de los métodos biométricos más confiables y por tanto más usados en el mundo, lo que implica que ya existe un amplio rango de tecnologías y técnicas diseñadas para el estudio de los ojos mediante imageneología, abriendo la puerta a otro tipo de estudios a partir de las mismas técnicas, como la relación entre enfermedades humanas y como estas pueden afectar la apariencia del iris de la persona afectada.
```python
### UTILIDADES PARA LA LECTURA DE IMÁGENES
%matplotlib inline

import cv2
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

def img_read(filename, mode = 'color'):
    if(mode == 'color'):   
        return cv2.cvtColor(cv2.imread(filename),cv2.COLOR_BGR2RGB)
    elif(mode=='grayscale' or mode=='greyscale' or mode == 'gray' or mode == 'grey'):
        return cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
    else:
        return None
    
def img_scale(img, value_range = [0,255]):
    return (value_range[1] - value_range[0])*(img - np.min(img))/(np.max(img)-np.min(img)) + value_range[0]

def img_trim(img, value_range = [0, 255]):
    res = img.copy()
    res[res > 255] = 255
    res[res < 0] = 0
    return res
```

#### 5.1 Experimentación en espacio de color RGB.
En pro de encontrar modificaciones útiles a nuestro fin decidimos utilizar configuraciones básicas de color en el espacio de color RGB, a continuación se tiene una tabla resumen de dicho proceso. Los valores que tienen una “X” no fueron afectados y tienen el valor que se leyó de la imágen inicial.

```python
def anular_canales(im, canales):
    
    for i in range(len(im)):
        for j in range(len(im[0])):
            if canales[0] == 0:
                im[i][j][0] = 0
            elif canales[0] == 1:
                im[i][j][0] = 255
            elif canales[0] == "x":
                break
                
            if canales[1] == 0:
                im[i][j][1] = 0
            elif canales[1] == 1:
                im[i][j][1] = 255
            elif canales[0] == "x":
                break
                
            if canales[2] == 0:
                im[i][j][2] = 0
            elif canales[2] == 1:
                im[i][j][2] = 255
            elif canales[0] == "x":
                break
                
    im2 = cv2.cvtColor(im, cv2.COLOR_RGB2BGR)
    return im2
          
#plt.imshow(anular_canales(img3.copy(), ["x",0,0]))
```

#### 5.3 Thresholding.
Realizando una inspección visual y una discusión entre el equipo, descubrimos que, en general, se presenta un alto contraste entre el iris y la pupila, a su vez también entre el iris y la periferia de este. Con el fin de explotar esta característica decidimos utilizar un umbral (threshold) para desaparecer el iris completamente y encontrar sólo los bordes interior y exterior del mismo.

```python
imag = img2.copy()
imag2 = img5.copy()

def thres(im, threshold):
    for i in range(len(im)):
        for j in range(len(im[0])):
            if im[i][j][0] > threshold:
                im[i][j][0] = 255
            else:
                im[i][j][0] = 0
                im[i][j][1] = 0
                im[i][j][2] = 0

    imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)

    for i in range(len(imgray)):
        for j in range(len(imgray[0])):
            if imgray[i][j] < threshold:
                imgray[i][j] = 255
            else:
                imgray[i][j] = 0
    
    return imgray

            
imag = thres(imag, 120)
imag2 = thres(imag2, 120)
    
fig, arreglo_plots2 = plt.subplots(1,2, figsize=(17, 10), squeeze=False)

arreglo_plots2[0][0].set_title('Ojo 2')
arreglo_plots2[0][0].imshow(imag, cmap="gray", vmin=0, vmax=255)

arreglo_plots2[0][1].set_title('Ojo 5')
arreglo_plots2[0][1].imshow(imag2, cmap="gray", vmin=0, vmax=255)
```

### 6. Segmentación.


#### Thresholding adaptativo y gaussiano
Para el proceso de segmentación de los diferentes elementos del iris usamos un thresholding adaptativo, es decir, un thresholding cuyo umbral se calcula respecto a pequeños deltas de la imagen lo que le premite ser más preciso a la hora de crear los contrastes. Para este paso utilizamos dos tipos de thresholding: Adaptativo de media y adaptativo Gaussiano.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def runAThresh(image):
    
    img = image.copy()
    
    img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.medianBlur(img_grey,5)

    return cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)

def runAGauss(image):
    
    img = image.copy()
    
    img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img = cv2.medianBlur(img_grey,5)

    return cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)


fig, (ax1,ax2) = plt.subplots(1,2, figsize=(20, 8))
ax1.set_title('Threshold adaptativo de media')
ax1.imshow(runAThresh(img_read('Fotos/17.jpg', 'color')), cmap='gray')
ax2.set_title('Threshold adaptativo gaussiano')
ax2.imshow(runAGauss(img_read('Fotos/17.jpg', 'color')), cmap='gray')
```

#### Reconocimiento de bordes: Sobel

Ahora que tenemos la imagen con altos contrastes en el iris podemos utilizar una técnica de detección de bordes que nos permita tener un mayor acercamiento a los patrones que se puedan presentar en el iris del ojo. A continuación presentamos los resultados:

```python
from skimage.filters import sobel_v,sobel_h, sobel
import skimage 

def runSobel(image):
    
    img = image.copy()
    
    gradiente_x = cv2.Sobel(img, ddepth=-100, dx = 1, dy = 0, ksize = 7)
    gradiente_y = cv2.Sobel(img, ddepth=-100, dx = 0, dy = 1, ksize = 7)

    magnitud_sobel = np.sqrt(np.power(gradiente_x,2)+np.power(gradiente_y,2))
    magnitud_sobel = img_scale(magnitud_sobel)

    return magnitud_sobel
    
#cv2.imwrite('sobel.png',magnitud_sobel)

fig, (ax1,ax2) = plt.subplots(1,2, figsize=(20, 8))
ax1.imshow(runSobel(img_read('Fotos/5.jpg', "gray").astype(np.float32)), cmap='gray')
ax2.imshow(runSobel(img_read('Fotos/15.jpg', "gray").astype(np.float32)))

matriz = runSobel(img_read('Fotos/15.jpg', "gray").astype(np.float32))

sum=0
count=0

for i in range(len(matriz)):
    for j in range(len(matriz[i])):
        sum = sum + matriz[i][j]
        count = count + 1
        
print(sum/count)
```

```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import data, color
from skimage.transform import hough_circle, hough_circle_peaks
from skimage.feature import canny
from skimage.draw import circle_perimeter
from skimage.util import img_as_ubyte


# Load picture and detect edges

image = img_as_ubyte(runAThresh(img_read('Fotos/5.jpg', 'color')))
edges = canny(image, sigma=3, low_threshold=10, high_threshold=50)


# Detect two radii
hough_radii = np.arange(20, 50, 2)
hough_res = hough_circle(edges, hough_radii)

# Select the most prominent 5 circles
accums, cx, cy, radii = hough_circle_peaks(hough_res, hough_radii,
                                           total_num_peaks=3)

# Draw them
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(10, 4))
image = color.gray2rgb(image)
for center_y, center_x, radius in zip(cy, cx, radii):
    circy, circx = circle_perimeter(center_y, center_x, radius)
    image[circy, circx] = (220, 20, 20)

ax.imshow(image, cmap=plt.cm.gray)
plt.show()
```

```python
import numpy as np
import cv2
import math
from matplotlib import pyplot as plt


def findPupil(image): 
    #Encontrar pupila con Hughes
    img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    img = cv2.medianBlur(img,5)
    cimg = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,1,20,500, param1=50,param2=70,minRadius=0,maxRadius=0)
    circles = np.uint16(np.around(circles))
    for i in circles[0,:]:
        # draw the outer circle
        cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)
        # draw the center of the circle
        cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)

    sum = 0
    counter = 0
    mini = 255
        
    #Recorrer resultados de Hughes y encontrar la pupila entre falsos positivos
    for i in circles[0,:]:
        for j in range(i[1]-i[2], min(i[1]+i[2],len(img))):
            for k in range(i[0]-i[2], min(i[0]+i[2],len(img[0]))):
                if math.sqrt( (j-i[1])*(j-i[1]) + (k-i[0])*(k-i[0]) ) <= i[2] :
                    sum = sum + img[j][k]
                    counter = counter + 1
        if counter > 0:
            mean = sum/counter
        
        if mean < mini:
            mini = mean
            x=i[0]
            y=i[1]
            r=i[2]
            
        sum = 0
        counter = 0
    return [img,x,y,r,cimg]
            
imagen_usar = 'Fotos/7.jpg'

newimg,x,y,r,newimgu = findPupil(img_read(imagen_usar,'color'))

cimg2 = cv2.cvtColor(newimg, cv2.COLOR_GRAY2BGR)
cv2.circle(cimg2,(x,y),r,(0,255,0),2)
cv2.circle(cimg2,(x,y),2,(0,0,255),3)

#Encontrar Iris
count = 0
ratio = 0
for i in range(x):
    if newimg[x - i][y] < 150:
        count = count + 1
        
        if count == 3:
            ratio = x - i
            break
    else:
        count = 0
        

imgv2 = img_read(imagen_usar,'gray')   

#Pintar de blanco las áreas que sobran
for j in range(y-r, min(y+r,len(newimg))):
    for k in range(x-r, min(x+r,len(newimg[0]))):
        if (math.sqrt( (j-y)*(j-y) + (k-x)*(k-x) ) <= r):
            imgv2[j][k]= 255
            
for j in range(len(newimg)):
    for k in range(len(newimg[0])):
        if (math.sqrt( (j-y)*(j-y) + (k-x)*(k-x) ) >= ratio):
            imgv2[j][k]= 255

imsobel = runSobel(imgv2.astype(np.float32))        
cv2.imwrite("Sobelpupil.png", imsobel)
            
#print(ratio)
fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(20, 8))
ax1.imshow(newimgu)
ax2.imshow(imgv2, cmap="gray")

count2 = 0
for i in range(len(imsobel)):
    for j in range(len(imsobel[i])):
        if imsobel[i][j] > 30:
            imsobel[i][j] = 255
            count2 = count2 + 1
            
ax3.imshow(imsobel, cmap="gray")

sum=0
count=0

for i in range(len(imsobel)):
    for j in range(len(imsobel[i])):
        sum = sum + imsobel[i][j]
        count = count + 1

        
print("Media: ", sum/count)
print("Porcentaje: ", count2/(len(imsobel)*len(imsobel[i])))
```
